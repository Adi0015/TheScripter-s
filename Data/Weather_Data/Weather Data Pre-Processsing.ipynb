```python
import pandas as pd
import numpy as np

data = pd.read_csv('Weather_Data.csv')
```

```python
data['District'].unique()
```

```python
districts = ['Adilabad','Warangal (R)','Nizamabad','Khammam','Warangal (U)','Karimnagar','Warangal Rural','Warangal Urban','Warangal']
```

```python
data_sorted = data[data.District.isin(districts)]
```

```python
data_sorted
```

```python
grouped_data = data_sorted.groupby(['Date','Mandal','District']).mean()
# Reset the index of the grouped data to create a new dataframe with one row per date for each location
data_final = grouped_data.reset_index()
```

```python
data_final.info()
```

```python
data_final['Date'] = pd.to_datetime(data_final.Date)
```
